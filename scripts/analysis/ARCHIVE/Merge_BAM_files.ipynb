```python
import pysam
```

```python
# Ingest bam files
all_context_file = pysam.AlignmentFile("/Data1/seq_data/TubeD1a_N2_Fiberseq_Hia5_MSssI_12_22_22/basecalls/m6A/mod_mappings.commonsorted2.bam",
                                "rb")
mC_file = pysam.AlignmentFile("/Data1/seq_data/TubeD1a_N2_Fiberseq_Hia5_MSssI_12_22_22/basecalls/5mC/mod_mappings.commonsorted2.bam",
                                "rb")

# Create sets to store the query names of the reads in the first BAM file
all_context_reads_set = set()
mC_reads_set = set()
common_all_context = []
common_5mC = []
```

```python
# COUNT NUMBER OF READS AND COMMON READS IN EACH INPUT FILE
for read in all_context_file.fetch():
    all_context_reads_set.add(read.query_name)

for read in mC_file.fetch():
    mC_reads_set.add(read.query_name)
    if read.query_name in all_context_reads_set:
        common_5mC.append(read)

for read in all_context_file.fetch():
    if read.query_name in mC_reads_set:
        common_all_context.append(read)

print ("READS IN ALL CONTEXT:",len(all_context_reads_set))
print ("READS IN 5mC:",len(mC_reads_set))
print ("READS IN BOTH:",len(common_all_context))
```

```python
### IF DIFFERING NUMBER OF READS, RECOMMEND CREATING NEW BAM FILES
# Open a new BAM file for writing
output_all_context_bam = pysam.AlignmentFile("/Data1/seq_data/TubeD1a_N2_Fiberseq_Hia5_MSssI_12_22_22/basecalls/m6A/mod_mappings.commonsorted.bam", "wb", template=all_context_file)
output_5mC_bam = pysam.AlignmentFile("/Data1/seq_data/TubeD1a_N2_Fiberseq_Hia5_MSssI_12_22_22/basecalls/5mC/mod_mappings.commonsorted.bam", "wb", template=mC_file)

# Write the reads that are present in both BAM files to the new BAM file
for read in common_all_context:
    output_all_context_bam.write(read)

for read in common_5mC:
    output_5mC_bam.write(read)

# Close the BAM files
output_all_context_bam.close()
output_5mC_bam.close()

# Sort the BAM file
# pysam.sort("/Data1/seq_data/TubeD1a_N2_Fiberseq_Hia5_MSssI_12_22_22/basecalls/m6A/mod_mappings.commonsorted.bam", "-o","/Data1/seq_data/TubeD1a_N2_Fiberseq_Hia5_MSssI_12_22_22/basecalls/m6A/mod_mappings.commonsorted2.bam")
# pysam.sort("/Data1/seq_data/TubeD1a_N2_Fiberseq_Hia5_MSssI_12_22_22/basecalls/5mC/mod_mappings.commonsorted.bam", "-o","/Data1/seq_data/TubeD1a_N2_Fiberseq_Hia5_MSssI_12_22_22/basecalls/5mC/mod_mappings.commonsorted2.bam")

# Index the BAM files
pysam.index("/Data1/seq_data/TubeD1a_N2_Fiberseq_Hia5_MSssI_12_22_22/basecalls/m6A/mod_mappings.commonsorted.bam")
pysam.index("/Data1/seq_data/TubeD1a_N2_Fiberseq_Hia5_MSssI_12_22_22/basecalls/5mC/mod_mappings.commonsorted.bam")

# Ingest bam files
all_context_file = pysam.AlignmentFile("/Data1/seq_data/TubeD1a_N2_Fiberseq_Hia5_MSssI_12_22_22/basecalls/m6A/mod_mappings.commonsorted.bam",
                                "rb")
mC_file = pysam.AlignmentFile("/Data1/seq_data/TubeD1a_N2_Fiberseq_Hia5_MSssI_12_22_22/basecalls/5mC/mod_mappings.commonsorted.bam",
                                "rb")
```
