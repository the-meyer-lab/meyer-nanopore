```python
import pandas as pd
import numpy as np
```

```python
### Goal is to add genes selected from groseq_fn to bed_fn based on criteria in escapee_selection_fn
bed_fn = "/Data1/reference/tss_tes_rex_combined_v6_WS235.bed"
escapee_selection_fn = "/Data1/reference/escapee_selection_v1.csv"
groseq_fn = "/Data1/reference/gro_seq_sdc2_rnai_kruesi.csv"

# Read all csvs into pandas df
bed_df = pd.read_csv(bed_fn, sep="\t")
escapee_selection_df = pd.read_csv(escapee_selection_fn, sep=",", index_col=0)
groseq_df = pd.read_csv(groseq_fn, sep=",")

# List of rows to convert to float
rows_to_convert = ["min fold change", "max fold change", "max p value",
                   "min p value", "max N2 norm count", "min N2 norm count"]

# Convert only the specified rows in escapee_selection_df to floats
for row in rows_to_convert:
    escapee_selection_df.loc[row] = escapee_selection_df.loc[row].astype(float)

# Display a few rows from each df
print(bed_df.columns)
display(bed_df.head())
print(escapee_selection_df.columns)
display(escapee_selection_df)
# print groseq_df columnnames
print(groseq_df.columns)
display(groseq_df.head())
```

```python
# Define a function to assign the 'type' based on the conditions.
def assign_type(row):
    matching_columns = []
    for column in escapee_selection_df.columns:
        min_fold_change = escapee_selection_df.loc["min fold change", column]
        max_fold_change = escapee_selection_df.loc["max fold change", column]
        min_p_value = escapee_selection_df.loc["min p value", column]
        max_p_value = escapee_selection_df.loc["max p value", column]
        min_count = escapee_selection_df.loc["min N2 norm count", column]
        max_count = escapee_selection_df.loc["max N2 norm count", column]

        # Check the conditions for the current column
        if (
            min_fold_change <= row["y93_N2_edgeRfoldChange"] <= max_fold_change
            and min_p_value <= row["adj_pValue"] <= max_p_value
            and min_count <= row["N2_RNAi_count"] <= max_count
        ):
            #print("FOUND MATCH!")
            matching_columns.append(column)

    #if matching_columns:
        #print("MATCHING COLUMNS: ", matching_columns)
    # If none of the conditions are met, return NaN
    return ";".join(matching_columns) if matching_columns else np.nan

# Apply the function to groseq_df
groseq_df['type'] = groseq_df.apply(assign_type, axis=1)

groseq_df_nonull = groseq_df[groseq_df['type'].notnull()]
# Display all groseq_df rows where type is not NaN
display(groseq_df_nonull)

#print columns
print(groseq_df_nonull.columns)

# Display all groseq_df rows where type has multiple matches
display(groseq_df_nonull[groseq_df_nonull['type'].str.contains(";")])
```

```python
# Filtering bed_df based on the type conditions
keywords = ["TSS", "TES", "gene"]
bed_df_filtered = bed_df[bed_df['type'].str.contains('|'.join(keywords), case=False, na=False)]
# Drop all rows containing "her" in type
bed_df_filtered = bed_df_filtered[~bed_df_filtered['type'].str.contains("her", case=False, na=False)]

# Filtering out rows with NaN in the 'type' column from groseq_df
groseq_df_filtered = groseq_df[groseq_df['type'].notna()]

# Collecting matching rows
new_rows = []

for _, row in bed_df_filtered.iterrows():
    mask = (
        (groseq_df_filtered['CHROMOSOME'] == row['chromosome']) &
        (
            (groseq_df_filtered['gene_start_WS235'] == row['start']) |
            (groseq_df_filtered['gene_end_WS235'] == row['start']) |
            (groseq_df_filtered['gene_start_WS235'] == row['end']) |
            (groseq_df_filtered['gene_end_WS235'] == row['end'])
        )
    )
    matching_groseq_rows = groseq_df_filtered[mask].copy()
    for _, match_row in matching_groseq_rows.iterrows():
        #print("MATCHING ROW: ", match_row)
        types = match_row['type'].split(';')
        for t in types:
            new_row = row.copy()
            new_row['type'] = row['type'] + ";" + t # Concatenating the types
            new_rows.append(new_row)

# Creating a dataframe from the new rows
new_rows_df = pd.DataFrame(new_rows)

# Replace all occurances of "_q1", "_q2", "_q3", "_q4" with ""
new_rows_df['type'] = new_rows_df['type'].str.replace("_q1", "").str.replace("_q2", "").str.replace("_q3", "").str.replace("_q4", "")

display(new_rows_df.head(100))
# Concatenate the original bed_df_filtered with the new rows
result_df = pd.concat([bed_df, new_rows_df], ignore_index=True)

# Displaying the first 100 rows of result_df
display(result_df.head(100))
```
